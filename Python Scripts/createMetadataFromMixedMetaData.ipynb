```python
merged_samples_df = pd.merge(metadata_df,transposed_df,left_on="Sample",right_on="Samples Transposed", how ="inner")
```

```python
#Sort by sample ID
merged_samples_df = merged_samples_df.sort_values(by = "Sample ID", ignore_index=True)
```

```python

# Function to generate new sample names
def generate_new_sample_name(row):
    # Extract the number after WOBEX using regex followed by a number, with or without separators
    match = re.search(r'WOBEX(?:[_-]?)(\d+)', row['Sample'])
    
    if match:
        number = match.group(1)
        # Concat W + number basically makes WOBEX3 or WOBEX_3 into W3
        new_name = f"W{number}"
        
        # Concat with Timepoint
        new_name += f"_{row['Timepoint']}"
        
        # Concat based on Pulldown column
        if row['Pulldown'] == 'GLAST':
            new_name += "_A"
        elif row['Pulldown'] == 'GluR2':
            new_name += "_R"
        else:
            print(f"Match not found: {row['Sample']}")
            return None
        
        # Concat the Sample ID w/o underscore
        new_name += str(row['Sample ID'])
        
        return new_name
    else:
        print(f"Match not found: {row['Sample']}")
        return None

# Apply the function to each row to create a new column
merged_samples_df['New Sample Names'] = merged_samples_df.apply(generate_new_sample_name, axis=1)

# Check if any samples could not be converted
if merged_samples_df['New Sample Names'].isnull().any():
    print("Some samples couldn't be converted.")
else:
    print("All samples converted successfully:")
    print(merged_samples_df[['Sample', 'New Sample Names']])
```
